```python

def foo(parent, child, dct):
    flag = False
    for i in range(0, len(dct[child])):
        sub_child = dct[child][i] # each parent in dct[child] list become child and check
        if (sub_child in dct) and (parent in dct[sub_child]):
            return True
        elif sub_child in dct:
            #print(sub_child)
            foo(parent, sub_child, dct)
        else:
            return False
    
```

```python
n = int(input())
dct = {}
for i in range(n):
    lst = []
    lst += input().split()
    if len(lst) == 1:
        key = str(lst[0])
        dct[key] = []
    else:
        key = str(lst[0])
        lst_for_dict = []
        for i in range(2, len(lst)):
            value = str(lst[i])
            lst_for_dict += value
        dct[key] = lst_for_dict
#print(dct)

m = int(input())
res_list = []
for i in range(m):
    lst = []
    lst += input().split()
    parent = str(lst[0])
    child = str(lst[1])
    if (parent == child) and (parent in dct) and (child in dct):
        res_list +=['Yes']
    elif (child in dct) and (parent in dct[child]):
        #print('Yes')
        res_list += ['Yes']
    elif (child in dct) and (parent not in dct[child]):
        res = foo(parent, child, dct)
        if res == True:
            res_list += ['Yes']
        else:
            res_list += ['No']
    else:
        res_list += ['No']
#print(res_list)
for i in res_list:
    print(i)
```
